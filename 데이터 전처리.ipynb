## 1.Data Set확인

***1-1.변수 확인***

###### - 그렇다면, 데이터의 종류에 대해 알아보자
###### <font color=yellow>* 정형 데이터</font>
######  - 구조화된 데이터를 의미 -> 열과 행으로 표시되는 테이블 형식 데이터(SQL문 사용)
######  - 테이블 형식의 데이터베이스를 '관계형 데이터베이스'라고 한다
######  - 정형 데이터는 데이터베이스의 정해진 규칙에 맞게 들어간 데이터 중 수치 만으로 의미 파악이 쉬운 데이터를 보통 뜻한다
######  - 즉, 값의 의미를 파악하기 쉽고, 규칙적인 값으로 데이터가 들어가는 경우 '정형 데이터'라고 할 수 있다!
###### 
***
###### <font color=yellow>* 비정형 데이터</font>
######  - 사전 정의된 방식으로 구성되지 않거나, 사전 정의된 데이터 모델이 없는 데이터를 뜻한다
######  - ex) 비디오, 오디오 및 바이너리(byte) 데이터 파일, 문서 파일, 텍스트, 음성, 영상
######  - '정해진 규칙이 없어 값의 의미를 쉽게 파악하기 힘든 데이터'가 비정형 데이터라고 생각하면 된다!
######
***
###### <font color=yellow>* 반정형 데이터</font>
######  - 구조화된 데이터(관계형 데이터베이스)로 구성되지 않지만, 여전히 어느 정도 구조를 가지고 있는 정보를 의미한다
######  - 정형 데이터와 같이 테이블의 행과 열로 구조화되어 있지는 않지만, 스키마 및 메타데이터 특성을 가지고 있다
######  - cf) 스키마 : 데이터베이스에서 자료의 구조, 자료의 표현 방법, 자료 간의 관계를 형식 언어로 정의한 구조
######  - cf) 메타 데이터 : 데이터에 관한 구조화된 데이터로, 다른 데이터를 설명해주는 데이터이다. 즉, 어떤 데이터 즉 구조화된 정보를 분석, 분류하고 부가적 정보를 추가하기 위해 그 데이터 뒤에 함께 따라가는 정보를 뜻한다
######  - ex) XML, HTML, JSON 등의 파일 형태로 저장


***1-2.RAW 데이터(미가공 데이터) 확인***

## 2.결측값 처리
###### - 결측값(Missing Value) : 필수적인 데이터가 입력되지 않고 누락된 값
###### - 결측값이 있는 상태로 모델을 만들면 변수간의 관계가 왜곡될 수 있기에 모델의 정확성이 떨어진다
###### - 결측값은 발생하는 유형이 다양하다, 결측값이 무작위로 발생하거나, 다른 변수와 관계가 있는지 여부에 따라 결측값을 처리하는 방법도 조금씩 달라진다

***결측값의 종류***
###### - 완전 무작위 결측 : 변수상에서 발생한 결측값이 다른 변수들과 아무런 상관이 없는 경우
###### - 무작위 결측 : 누락된 자료가 특정 변수와 관련되어 일어나지만, 그 변수의 결과는 관계가 없는 경우
###### - 비 무작위 결측 : 누락된 값이 다른 변수와 연관 있는 경우
***
***결측값 처리 방법의 종류***  

**2-1. 삭제**
###### - 전체 삭제 : 결측값이 발생한 모든 관측치를 삭제 -> 간편하지만, 관측치가 줄어들어 모델의 유효성이 떨어질 수 있다
###### - 부분 삭제 : 데이터 중 모델에 포함시킬 변수들 중 결측값이 발생한 관측치만 삭제 -> 모델에 따라 변수가 제각각 다르기 때문에 관리 Cost가 늘어난다는 단점이 있다  


***2-2. 다른 값으로 대체(평균, 최빈값, 중간값)***
###### - 결측값이 발생한 경우 다른 관측치의 평균, 최빈값, 중간값 등으로 대체할 수 있다
###### - 일괄 대체 방법 : 모든 관측치의 평균값 등으로 대체
###### - 유사 유형 대체 방법 : 범주형 변수를 활용하여 유사한 유형의 평균값 등으로 대체
###### - ex) 남자 키의 평균값 : 173 / 여자 키의 평균값 : 158인 경우, 남자 관측치의 결측값은 173으로 대체  

***2-3. 예측값 삽입***
###### - 결측값이 없는 관측치를 Train data로 사용하여 결측값을 예측하는 모델을 만든다
###### - 이 모델을 통해 결측값이 있는 관측 데이터의 결측값을 예측하는 방법 -> Regression이나 Logistic regression을 주로 사용한다
###### - 결측값을 다른 값으로 대체하는 방법보다는 덜 자의적이지만, 결측값이 다양한 변수에서 발생하는 경우 힘들고, 모델의 예측력이 낮은 경우에는 사용하기 어려운 방법이다

## 3.이상값 처리
###### - 이상값이란 데이터/샘플과 동떨어진 관측치로, 모델을 왜곡할 가능성이 있는 관측치를 의미한다
***이상값 찾아내기***
###### - 이상값을 찾아내기 위한 가장 쉬운 방법은 변수의 분포를 시각화하는 것이다
###### - 변수 1개 : Boxplot or Histogram / 변수 2개 : Scatter plot
![이상값 찾기](img/python2.png)

***이상값 처리하기***  

***3-1. 단순 삭제***
###### - 이상값이 Human error에 의해 발생한 경우에는 해당 관측치를 삭제하면 된다
###### - ex) 단순 오타, 주관식 설문 등의 비현실적인 응답  

***3-2. 다른 값으로 대체***
###### - 절대적인 관측치의 숫자가 작은 경우, 삭제의 방법으로 이상치를 제거하면 관측치의 절대량이 작아지는 문제가 발생한다
###### - 위의 경우, 이상값이 Human error에 의해 발생했더라도 관측치를 삭제하는 대신 평균과 같은 다른 값으로 대체하거나, 결측값과 유사한게 다른 변수들을 사용하여 예측 모델을 만들고, 이상값을 예측한 후 해당 값으로 대체하는 방법도 존재한다  

***3-3. 변수화***
###### - 이상값이 자연 발생한 경우, 단순 삭제나 대체의 방법을 통해 수립된 모델은 설명/예측하고자 하는 현상을 잘 설명하지 못할 수 있다
###### - 예를 들어, 아래 그래프에서 다른 관측치들만 보면 경력과 연봉이 비례하는 현상이 존재하는 것처럼 보이지만, 5년 차의 연봉 $35,000인 이상치가 포함됨으로써 모델의 설명력이 크게 낮아져버린다

![이상값](img/py.webp)

###### - 자연발생적인 이상값의 경우, 바로 삭제하지 말고 좀 더 찬찬히 이상값에 대해 파악하는 것이 중요하다
###### - 위의 이상값의 경우, 의사 등 전문직종에 종사하는 사람이라고 가정했을 때 전문직 종사 여부를 Yes-No로 변수화하면 이상값을 삭제하지 않고 모델에 포함시킬 수 있게 된다  

***3-4. 리샘플링***
###### - 자연 발생한 이상값을 처리하는 또 다른 방법으로 해당 이상값을 분리하여 모델을 만드는 방법이 있다
###### - 아래와 같이 15년 이상의 경력을 가진 이상값이 존재한다고 할 때, 이 관측치는 경력은 길지만 연봉이 비례해서 늘어나지 않은 경우이다

***3-5. 케이스를 분리하여 분석***
###### - 위와 동일한 사례에서 실은 경력이 지나치게 길어질 경우 연봉이 낮아지는 현상이 실제로 발생한 것일 수 있다(건강상의 이유 등으로..)

###### - 보다 좋은 방법은 이상값을 포함한 모델과 제외한 모델을 모두 만들고 각각의 모델에 대한 설명을 다는 것이다
###### - 자연 발생한 이상값에 별다른 특이점이 발견되지 않는다면, 단순제외 보다는 케이스를 분리하여 분석하는 것이 좋은 방법이다

## 4.Feature Engineering
###### - Feature Engineering이란, 기존의 변수를 사용하여 데이터에 정보를 축하는 일련의 과정이다
###### - 새로 관측치나 변수를 추가하지 않고도 기존의 데이터를 보다 유용하게 만드는 방법론이다

***4-1. SCALING***
###### - 변수의 단위를 변경하고 싶거나, 변수의 분포가 편향되어 있을 경우, 변수 간의 관계가 잘 드러나지 않는 경우에는 변수 변환의 방법을 사용한다
###### - 가장 자주 사용하는 방법으로는 Log함수가 있고, 유사하지만 좀 덜 자주 사용되는 Square root를 취하는 방법도 존재한다  

***4-2. BINNING***
###### - 연속형 변수를 범주형 변수로 만드는 방법이다
###### - 예를 들어, 연봉 데이터가 수치로 존재하는 경우, 이를 100만원 미만, 101만원~200만원.. 하는 식으로 범주형 변수로 변환하는 것이다
###### - Binning에는 특별한 원칙이 있는 것이 아니므로, 분석가의 Business이해도에 따라 창의적인 방법으로 Binning될 수 있다  

***4-3. TRANSFORM***
###### - 기존 존재하는 변수의 성질을 이용하여 다른 변수를 만드는 방법이다
###### - 예를 들어, 날짜 별 판매 데이터가 있다면, 날짜 변수를 주중/주말로 나눈 변수를 추가한다던지, eSports의 관람객 데이터의 경우 해당일에 SKT T1의 경기가 있는 여부 등을 추가하는 것이다
###### - Transform에도 특별한 원칙이 있는 것은 아니며, 분석가의 Business이해도에 따라 다양한 변수가 생성될 수 있다  

***4-4. DUMMY***
###### - Binning과 반대로 범주형 변수를 연속형 변수로 변환하기 위해 사용된다
###### - 사용하고자 하는 분석 방법론에서 필요한 경우에 주로 사용한다

***결국, Garbage-in, garbage-out이기 때문에 Model을 생성하기 전에 내가 가지고 있는 데이터의 상태를 확인하고, 내가 설계한 분석 방법에 맞는 형태로 적절하게 전처리해주는 것은 정확한 결과를 얻기 위한 필수적인 단계라고 할 수 있다...***
