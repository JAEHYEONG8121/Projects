```python
import os

# 파일 경로 설정
file_path = '/usr/local/lib/python3.10/dist-packages/TTS-0.0.9.2-py3.10-linux-x86_64.egg/TTS/tts/datasets/TTSDataset.py'

# 파일 내용 읽어오기
with open(file_path, 'r') as file:
    lines = file.readlines()

# collections.Mapping을 collections.abc.Mapping으로 변경
with open(file_path, 'w') as file:
    for line in lines:
        file.write(line.replace('collections.Mapping', 'collections.abc.Mapping'))
```

```python
import os

# 파일 경로 설정
file_path = '/usr/local/lib/python3.10/dist-packages/librosa/util/utils.py'

# 파일 내용 읽어오기
with open(file_path, 'r') as file:
    lines = file.readlines()

# np.float을 float으로 변경
with open(file_path, 'w') as file:
    for line in lines:
        file.write(line.replace('np.float', 'float').replace('np.complex', 'complex'))
```
